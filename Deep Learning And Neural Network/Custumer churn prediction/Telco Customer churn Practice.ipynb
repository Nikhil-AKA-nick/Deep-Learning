```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df = df.drop("customerID", axis = 1)
```

```python
def data(df):
    for col in df:
        print(f'{col} :  {df[col].unique()}')
```

```python
df = df.replace(["No phone service","No internet service"], "No") 
```

```python
data(df)
```

```python
pd.to_numeric(df.TotalCharges, errors= "coerce")
```

```python
df[pd.to_numeric(df.TotalCharges, errors="coerce").isnull()]
```

```python
df.shape
```

```python
df1 = df[df.TotalCharges != " "]
```

```python
df1.shape
```

```python
df1.dtypes
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
df1.dtypes
```

```python
df1.head()
```

```python
yes_no_column = ["Partner","Dependents","PhoneService","MultipleLines","OnlineSecurity","OnlineSecurity","DeviceProtection","TechSupport","StreamingTV","StreamingMovies","PaperlessBilling","Churn","OnlineBackup"]
```

```python
for col in yes_no_column:
    df1[col].replace({"Yes" : 1, "No" : 0}, inplace = True)
```

```python
data(df1)
```

```python
df1.head()
```

```python
df1 = df1.replace({"Female" : 1, "Male" : 0})
```

```python
df1.head()
```

```python
le_column = ["InternetService","Contract","PaymentMethod"]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for col in le_column:
    df1[col] = le.fit_transform(df1[col])
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
cols_to_scale = ["tenure","MonthlyCharges","TotalCharges"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
```

```python
df1.head()
```

```python
data(df1)
```

```python
X = df1.drop("Churn", axis = 1)
```

```python
y = df1["Churn"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X ,y , test_size= 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(19, input_shape = [19], activation = "relu"),
    keras.layers.Dense(19, activation = "relu"),
    keras.layers.Dense(15, activation = "relu"),
    keras.layers.Dense(10, activation = "relu"),
    keras.layers.Dense(7, activation = "relu"),
    keras.layers.Dense(1, activation = "sigmoid")
])

model.compile(
    optimizer = "adam",
    loss = "mse",
    metrics = ["accuracy"]
    )

model.fit(X_train, y_train, epochs = 500)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
